# importing library 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Loading and processing

```python
data= pd.read_csv("1.01. Simple linear regression.csv")

gpa_x=data['GPA'].to_numpy()
sat_y=data['SAT'].to_numpy()
```

# Plotting

```python
plt.scatter(gpa_x, sat_y)

plt.title("SAT vs GPA")
plt.xlabel("GPA")
plt.ylabel("SAT")
```

# Calculation 

```python
#m= (mean(x)*mean(y)-(mean(x*y)))/((mean(x)**2- mean(x**2)))
#c = mean(y)-m*mean(x)
#hypothesis == predicted_sat = m*X-c
x_mean= np.mean(gpa_x)
y_mean = np.mean(sat_y)
x_squre_mean = np.mean(gpa_x**2)
xy_mean= np.mean(gpa_x *sat_y)

m= ((x_mean*y_mean)-xy_mean) / (x_mean**2 - x_squre_mean)
c=y_mean-m*x_mean
```

#  Linear regression plot

```python
p_y=m*gpa_x + c
plt.plot(gpa_x, p_y, color='black')
plt.scatter(gpa_x, sat_y)

plt.title("SAT vs GPA")
plt.xlabel("GPA")
plt.ylabel("SAT")
```


# Prediction

```python
x=float(input("Enter gpa: "))
predicted_y = m*x + c
print(predicted_y)
```

# Accuracy Rate

```python
total_u=0
for i in range(len(p_y)):
    total_u = total_u + (p_y[i]-y_mean)**2

total_d=0
for x in range(len(sat_y)):
    total_d = total_d + (sat_y[x]-y_mean)**2

r_squre = (total_u/total_d)

print(r_squre)
```
